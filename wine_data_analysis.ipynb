```python
# DIFFERENT FLAVOURS OF WINE
flavours  = ['fruity','earthy','spicy','smoky','flowery','acidic']
```

```python
flavours
```

```python
count_flavour = []

for f in flavours:
    num = len(wine_df[wine_df['description'].str.contains(f.lower())])
    count_flavour.append(num)
    
```

```python
print(count_flavour)
```

```python
plt.bar(flavours,count_flavour)
plt.title("Flavours of Wine")
plt.xlabel("Flavours")
plt.ylabel("Quantity")
plt.show()
```
